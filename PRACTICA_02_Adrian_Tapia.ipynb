# Ejercicio 1
Crea un diccionario con la siguiente información de tu vida (edad, sexo, estudios, nombre_amigos) [ojo, *nombre_amigos* será una lista con varios nombres]

```python
mi_dic = {
    "edad": 22,
    "sexo": "hombre",
    "estudios": "posgrado",
    "nombre_amigos": ["Carlota", "Álex", "Gigi", "Lucía", "Nadia"]
}
print(mi_dic)
```

# Ejercicio 2
Crea una función que imprima el diccionario anterior

```python
def imprime_dic(diccionario):
  for clave, valor in diccionario.items():
    print(f"{clave}: {valor}")

imprime_dic(mi_dic)
```

# Ejercicio 3
Crea una clase que reciba como entrada el diccionario y lo imprima

```python
class Diccionario:
  def __init__(self, diccionario):
    self.diccionario = diccionario
  def imprime_dic (self):
    for clave, valor in self.diccionario.items():
      print(f"{clave}: {valor}")
d1 = Diccionario(mi_dic)
d1.imprime_dic()
```

# Ejercicio 4
Crea una lista de diccionarios con distintas personas como clase

```python
# Se crea una clase que admite como entrada los argumentos de los ejercicios anteriores.
# Esta clase dará como resultado un diccionario con los argumentos de entrada como información de cada persona.
class Persona:
  def __init__(self, edad, sexo, estudios, nombre_amigos):
    self.edad = edad
    self.sexo = sexo
    self.estudios = estudios
    self.nombre_amigos = nombre_amigos
  def diccionario (self):
    return {
        "edad": self.edad,
        "sexo": self.sexo,
        "estudios": self.estudios,
        "nombre_amigos": self.nombre_amigos
    }

# Se definen distintas personas en función de los atributos definidos en la clase
# Estas personas se guardan en los objetos p1, p2 y p3
p1 = Persona(25, "hombre", "grado universitario", ["Antonio", "Luisa"])
p2 = Persona(27, "mujer", "posgrado", ["Maite", "Irene", "Juan"])
p3 = Persona(35, "hombre", "grado medio", ["Víctor", "Edurne"])

# Se crea una lista en la que se guardan los diccionarios de la información de vida de estas personas guardadas en los objetos p1, p2 y p3
lista = [
    p1.diccionario(),
    p2.diccionario(),
    p3.diccionario()
]
```

# Ejercicio 5
Imprime la lista anterior con un for()

```python
for persona in lista:
  print("\n") # Cada vez que recorra un elemento del objeto "lista" salta una línea
  for clave, valor in persona.items():
        print(f"{clave}: {valor}") # Por cada par clave, valor que se encuentre en la lista de items de los elementos "persona" se imprime la clave y el valor
```

# Ejercicio 7
Genera una variable Y que sea una distribución normal(0, 1)

```python
import numpy as np

Y = ((1 / (np.sqrt(2 * np.pi))) * np.exp((-X**2)/2))
# La variable Y se define con la fórmula de la distribución normal(0, 1)
# En función de la variable X creda anteriormente

print(Y)
```

# Ejercicio 8

```python
import matplotlib.pyplot as plt

plt.plot(X, Y, color="black")
plt.title("Curva Normal")
plt.xlabel("Variable X")
plt.ylabel("Densidad de probabilidad")
plt.show
```

# Ejercicio 9
Crea un vector que vaya de 0 a 10, conviértelo en una matriz de 2x5, multiplícalo por una matriz de unos y el resultado conviértelo en un tensor.

```python
vector = np.arange(0, 10)
print("El vector es:", "\n", vector)

reshaped = vector.reshape((2, 5))
print("El vector cambiado de forma es:", "\n", reshaped)

unos = np.ones((5, 2))
print("La matriz de unos es:","\n", unos)

multiplicacion = np.matmul(unos, reshaped)
print("La matriz resultante de la multiplicacion es:","\n", multiplicacion)

import torch
tensor = torch.from_numpy(multiplicacion)
print("El tensor final es:", "\n",tensor)
```

# Ejercicio 10
Calcula la derivada de x**3-2*x y dibuja ambas

```python
import torch
import matplotlib.pyplot as plt
import numpy as np

x = torch.linspace(-3, 3, 600, requires_grad = True)
print("El tensor de x es:", "\n", x)

y = x*3-2*x

y.backward(torch.ones_like(x))

x_vals = x.detach().numpy()
y_vals = y.detach().numpy()
dy_dx_vals = x.grad.detach().numpy()

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label="$y = x*3 - 2*x$", color="blue")
plt.plot(x_vals, dy_dx_vals, label="Derivada $dy/dx$", color="red", linestyle="--")
plt.legend()
plt.grid(True)
plt.show()
```
